**For third output**  
Create a dataset that shows which customers have yet to book a flight with Prep Air in 2024  
Create a field which will allow Prep Air to see how many days it has been since the customer last flew (compared to 31/01/2024  )
Categorise customers into the following group  s:
Recent fliers - flown within the last 3 mon  ths
Taking a break - 3-6 months since last fl  ight
Been away a while - 6-9 months since last f  light
Lapsed Customers - over 9 months since last flight

```python
# Joining Customer and Ticket table.

YetToBook_df = pd.merge(customer_df, ticket_df, how = 'left', on = ['Customer ID'])
```

```python
YetToBook_df.head()
```

```python
#Getting the values that are not in the joined table fo required output.

yettobook_df_final = YetToBook_df[YetToBook_df['Ticket Price'].isnull()]
```

```python
yettobook_df_final
```

```python
# Adding new column to the table with default value.

yettobook_df_final['Yet to book as of'] = '31/01/2024'
```

```python
yettobook_df_final
```

```python
# Converting both the columns (YetToBookAsOf & LastDateFlown) to date data type.

yettobook_df_final['Yet to book as of'] = pd.to_datetime(yettobook_df_final['Yet to book as of'])
yettobook_df_final['Last Date Flown'] = pd.to_datetime(yettobook_df_final['Last Date Flown'])
```

```python
# Getting the difference in number of days format and storing in new column.

yettobook_df_final['Days Since Last Flown'] = (yettobook_df_final['Yet to book as of'] - yettobook_df_final['Last Date Flown']).dt.days
```

```python
yettobook_df_final
```

```python
#Dropping the extra columns.

Third_Output_df = yettobook_df_final.drop(['Date', 'Flight Number', 'Ticket Price', 'Yet to book as of'], axis=1)
```

```python
Third_Output_df
```

```python
#Creating a function to add a category to all the customers based on the number of days since last flown and storing it in a new column.

def categorize_customer(days):
    if days <= 90:
        return 'Recent Flyer'
    elif days <= 180:
        return 'Taking a break'
    elif days <= 270:
        return 'Been away a while'
    else:
        return 'Lapsed Customers'

Third_Output_df['Customer Category'] = Third_Output_df['Days Since Last Flown'].apply(categorize_customer)
```
